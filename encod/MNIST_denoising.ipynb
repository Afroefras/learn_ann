## Import

```python
BASE_DIR = ''
```

```python
from keras.datasets import mnist
import numpy as np

#Load the MNIST data, don't need the labels just the images
(X_train, _), (X_test, _) = mnist.load_data()
#Normalize all pixels from 0-1
X_train = X_train.astype('float32') / 255.
X_test = X_test.astype('float32') / 255.
```

## Functions

```python
import time
import numpy as np
from IPython.lib.display import Audio

start = time.time()
def time_exp(x):
    minutes, seconds = np.floor(x/60), 60*(x/60-np.floor(x/60))
    print(f"{'{:.0f}'.format(minutes)} minutos con {'{:.2f}'.format(seconds)} segundos")
    
def tono(a = 1000, b = 700, play_time_seconds = 1, framerate = 4410):
    t = np.linspace(0, play_time_seconds, framerate*play_time_seconds)*np.pi
    return Audio(np.sin(a*t) + np.sin(b*t), rate = framerate, autoplay=True)
```

```python
def artif_noise(X,noise=0.5):
    #Recreate random noise in order to train the model "how to clean it"
    return np.clip(X+noise*np.random.normal(0.,1.,size=X.shape),0.,1.)
```

```python
from tensorflow.keras import Input,Model
from tensorflow.keras.layers import Conv2D,MaxPooling2D,UpSampling2D

def autoencod_builder(input_shape,layers):
    input_img = Input(shape=input_shape)
    x = input_img
    #Encoding-phase
    for layer in layers:
        x = Conv2D(filters=layer, kernel_size=3, activation='relu', padding='same')(x)
        x = MaxPooling2D(pool_size=2, padding='same')(x)
    #Decoding-phase
    for layer in layers[::-1]: #From the end to the beginning, to mantain symmetry
        x = Conv2D(layer, 3, activation='relu', padding='same')(x)
        x = UpSampling2D(2)(x)
    decoded = Conv2D(1, 3, activation='sigmoid', padding='same')(x)
    autoencod = Model(input_img, decoded)
    print(autoencod.summary())
    return autoencod
```

```python
import random
import matplotlib.pyplot as plt

def clean_noise(autoencod,noisy,n=7):
    decoded_imgs = autoencod.predict(noisy)
    plt.figure(figsize=(n*2, n//2))
    for i in range(1,n+1):
        r = random.randint(0,len(noisy))
        #This is how noising looks like!
        ax = plt.subplot(2,n,i)
        plt.imshow(noisy[r].reshape(28,28))
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)

        #Display reconstruction on X_test
        ax = plt.subplot(2,n,i+n)
        plt.imshow(decoded_imgs[r].reshape(28,28))
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
    plt.show()
```

## Artificial noise

```python
X_train_n = artif_noise(X_train)
X_test_n = artif_noise(X_test)
```

## Architecture

```python
autoencoder = autoencod_builder((28,28,1),[64,32])
```

## Training

```python
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
autoencoder.fit(X_train_n, X_train,
                epochs=1,
                batch_size=128,
                shuffle=True,
                validation_data=(X_test_n, X_test))
```

## Save the model

```python
import os
import pickle

#First, save the architecture
with open(os.path.join(BASE_DIR,'cnn_denoise_architecture.pickle'), 'w') as f:
    f.write(autoencoder.to_json())

#Then, the paramaters learned by itself
autoencoder.save_weights(os.path.join(BASE_DIR,'cnn_denoise_weights.h5'))
```

```python
import os
import pickle
from keras.models import model_from_json

with open(os.path.join(BASE_DIR,'cnn_denoise_architecture.pickle'), 'r') as f:
    model = model_from_json(f.read())

#Now, we can use this model whenever we want!
model.load_weights(os.path.join(BASE_DIR,'cnn_denoise_weights.h5'))
model.compile(optimizer='adam', loss='binary_crossentropy')
```

## Cleaning

```python
clean_noise(model,X_test_n)
```
